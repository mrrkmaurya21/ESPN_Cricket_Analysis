```python
import pandas as pd
cricket_data1=pd.read_csv("WC_AUS_Innings.csv")
```

```python
def calculate_batsman_performance(cricket_data):
    import pandas as pd
    # Filter data for the chosen batsman
    chosen_batsman = input("Please Enter Player Name: ")
    batsman_data = cricket_data[cricket_data['batsman'] == chosen_batsman].copy()

    # Calculate 'extra_balls' and 'valid_balls' based on 'ball_type'
    batsman_data['extra_balls'] = (batsman_data['ball_type'] != 'valid').astype(int)
    batsman_data['extra_runs'] = (batsman_data[batsman_data['ball_type'] != 'valid']['run']).astype(int)
    batsman_data['valid_balls'] = (batsman_data['ball_type'] == 'valid').astype(int)
    batsman_data['dot_balls'] = (batsman_data['run'] == 0).astype(int)

    # Group data by bowler and calculate performance metrics
    bowler_performance = batsman_data.groupby('bowler').agg({
        'run': 'sum',  # Calculate total runs scored against each bowler
        'extra_runs': 'sum',  # Calculate total extra runs scored against each bowler
        'dot_balls': 'sum',  # Calculate dot balls against each bowler
        'extra_balls': 'sum',  # Calculate extra balls against each bowler
        'valid_balls': 'sum',  # Calculate valid balls against each bowler
    }).reset_index()

    # Rename columns for clarity
    bowler_performance.columns = ['Bowler', 'TotalRuns', 'ExtraRuns', 'DotBalls', 'ExtraBalls', 'TotalBallsFaced']
    # Convert 'ExtraRuns' column to integers
    bowler_performance['ExtraRuns'] = bowler_performance['ExtraRuns'].astype(int)

    # Calculate Strike Rate and format it to 2 decimal places
    bowler_performance['StrikeRate'] = ((bowler_performance['TotalRuns'] / bowler_performance['TotalBallsFaced']) * 100).round(2)
    print()
    print(f"Performance of {chosen_batsman} against bowlers:")
    # Display the performance metrics for the chosen batsman against each bowler
    return bowler_performance
```

```python
# Assuming you have your cricket data loaded in a DataFrame called 'cricket_data'
calculate_batsman_performance(cricket_data1)
```
